```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_upper = [int(0), int(1), int(2), int(3), int(4)]
x_lower = [int(2), int(3), int(4)]

y_upper = [28, 28, 28, 28, 17]
y_lower = [11, 15, 17]
```

```python
#=============================================================================#
#                                  Plotting                                   #
#=============================================================================#
def set_size(width,fraction=1, subplots=(1, 1), ratio=0):
    """
    Source: https://jwalton.info/Embed-Publication-Matplotlib-Latex/
    
    Set figure dimensions to avoid scaling in LaTeX.
    param width: float or string
            Document width in points, or string of predined document type
    param fraction: float, optional
            Fraction of the width which you wish the figure to occupy
    param subplots: array-like, optional
            The number of rows and columns of subplots.
    returns:
    fig_dim: tuple
            Dimensions of figure in inches
    """
    if width == 'thesis':
        width_pt = 426.79135
    elif width == 'beamer':
        width_pt = 307.28987
    else:
        width_pt = width

    # Width of figure (in pts)
    fig_width_pt = width_pt * fraction
    # Convert from pt to inches
    inches_per_pt = 1 / 72.27

    # Golden ratio to set aesthetic figure height
    # https://disq.us/p/2940ij3
    if ratio == 0:
        golden_ratio = (5**.5 - 1) / 2
    else:
        golden_ratio = ratio  

    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_width_in * golden_ratio * (subplots[0] / subplots[1])

    return (fig_width_in, fig_height_in)

# Latex Font See Ref. https://matplotlib.org/stable/tutorials/text/usetex.html
tex_fonts = {
    # Use LaTeX to write all text
    "text.usetex": True,
    "font.family": "serif",
    "font.sans-serif": "Palantino",
    # Use 10pt font in plots, to match 10pt font in document
    "axes.labelsize": 12,
    "font.size": 12,
    # Make the legend/label fonts a little smaller
    "legend.fontsize": 12,
    "xtick.labelsize": 12,
    "ytick.labelsize": 12
}

plt.style.use('default')
plt.rcParams.update(tex_fonts)

width = 'thesis'
fig, ax = plt.subplots(ncols=1)
fig.tight_layout(pad=1.5)


# Set names and Labels
ax.set_xlabel('Benders Iteration')
ax.set_ylabel('Cost function [a.u.]')


ax.plot(x_upper,y_upper, 'ro', linestyle = 'dashed', label=r'$f(\mathbf(x))$')
ax.plot(x_lower, y_lower, 'bx', linestyle = 'dashed', label=r'$q(\mu, \lambda)$')
ax.grid()
ax.legend()

plt.savefig('LowerUpperBounds.pdf',  bbox_inches='tight')  
plt.show()
```
